```python
import pandas as pd
import numpy as np
```

```python
book=pd.read_csv('book.csv',encoding='latin-1')
book
```

```python
book_1=book.drop('Unnamed: 0',axis=1)
book_1
```

```python
len(book_1['Book.Title'].unique())
```

```python
books_df =book_1.pivot_table(index='User.ID',
                                 columns='Book.Title',
                                 values='Book.Rating').reset_index()
books_df
```

```python
books_df.index = book_1['User.ID'].unique()
```

```python
books_df
```

```python
books_df.fillna(0, inplace=True)
```

```python
books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
#from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( books_df.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = books_df['User.ID'].unique()
user_sim_df.columns = books_df['User.ID'].unique()
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
